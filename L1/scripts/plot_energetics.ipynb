```python
import numpy as np
import xarray as xr
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from latex import set_size


plt.style.use("tex")
sns.set_palette("colorblind")

figdir = "../figures/"
datadir = "../data/"
```

```python
data1 = pd.read_csv(datadir + "energy_f1.dat",
                   delim_whitespace=True,
                   index_col=0,
                   skipinitialspace=True,
                   names=["time", "potential", "kinetic", "total"]
                  )
data2 = pd.read_csv(datadir + "energy_f2.dat",
                   delim_whitespace=True,
                   index_col=0,
                   skipinitialspace=True,
                   names=["time", "potential", "kinetic", "total"]
                  )
```

```python
fig, ax = plt.subplots(2, 1, figsize=set_size(width=221, subplots=(2, 1)), sharex=True, sharey=True)
ax[0].plot(data1.index, data1.potential/data1.total)
ax[0].plot(data1.index, data1.kinetic/data1.total)
ax[0].axhline(1/3,
           color="black",
           linewidth=2
          )
ax[0].axhline(2/3,
           color="black",
           linewidth=2
          )
ax[0].set_ylabel(r"Relative energy [ ]")

ax[1].plot(data2.index, data2.potential/data2.total)
ax[1].plot(data2.index, data2.kinetic/data2.total)
ax[1].axhline(1/3,
           color="black",
           linewidth=2
          )
ax[1].axhline(2/3,
           color="black",
           linewidth=2
          )
ax[1].set_xlabel(r"Time index [ ]")
ax[1].set_ylabel(r"Relative energy [ ]")

sns.despine(fig=fig)
fig.tight_layout()
#fig.savefig(figdir + "energetics_f.pdf", format="pdf", bbox_inches="tight")
```
